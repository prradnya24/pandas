Case Study 1: Creating a Series of Student Scores 
Scenario: A teacher wants to store scores of 5 students in a subject. 
Input: 
python 
import pandas as pd 
scores = pd.Series([85, 90, 78, 92, 88], index=['Amit', 'Bhavna', 'Chetan', 'Divya', 'Esha']) 
print(scores) 
Expected Output: 
Code 
Amit     
 85 
Bhavna    90 
Chetan    78 
Divya     92 
Esha      88 
dtype: int64

```python
import pandas as pd 
scores=[85,90,78,92,88]
index=['amit','bhavana','chetan','divya','esha']

scores = pd.Series(scores,index=index) 
print(scores) 
```

Case Study 6: Data Alignment Between Two DataFrames

science = pd.Series([90, 82, 95], index=['Bhavna', 'Chetan', 'Divya']) 
aligned = pd.DataFrame({'Math': math, 'Science': science}) 
print(aligned) 
Expected Output: 
Code 
Math  Science

Bhavna  78.0     90.0 
Chetan  92.0     82.0 
Divya    NaN     95.0
